# Random Forest Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('kolkata_updated_dec28th_copy2.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```





With np.isnan(X) you get a boolean mask back with True for positions containing NaNs

```
np.isnan(x)
```

With np.where(np.isnan(X)) you get back a tuple with i, j coordinates of NaN

```
np.where(np.isnan(x))
```

with np.nan_to_num(X) you "replace nan with zero and inf with finite numbers"

```
np.nan_to_num(x)
```

## Training the Random Forest Regression model on the whole dataset

```
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(x,y)
```

## Predicting a new result

```
regressor.predict([[70]])
```

## Visualising the Random Forest Regression results (higher resolution)

```
X_grid = np.arange(min(x), max(x), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title(' PM 2.5 vs Air Quality Index (Random Forest Regression)')
plt.xlabel('PM 2.5 levels')
plt.ylabel('AQI')
plt.show()
```
